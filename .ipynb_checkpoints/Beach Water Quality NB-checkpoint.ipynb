## BEACH WATER QUALITY

The city of Chicago, on Lake Michigan,in Iliinois has approximately 26 miles of free and open lakefront for leisure activities of Chicago residents and visitors. The Chicago Park District manages the city's lakefront. 

## Business Understanding:

The stakeholder, Surf Club, is a social club of fun-loving beachgoers and surfers. They are based in Chicago. As a club, they face one major challenge: swimming bans placed on the beaches by the regulatory agency due to the risk of high bacterial content in the beach waters leading to bacterial poisoning.  
Before now, swimming bans during the beach season along Lake Michigan's shores were a major, and all too frequent, frustration for Chicago beach-goers, like members of the Surf Club. Swimming bans would be enforced when daily water samples showed very high levels of E. coli, which in itself is not the main danger, but signals that the water could be contaminated with other, more dangerous bacteria. Time lags between sampling and testing often meant that the beaches were open to swimmers while the bacterial levels were high and by the time the results came in and bans were in place, bacterial levels could have gone down. The Chicago Park District has moved away from individual samples in favor of buoys and weather stations that transmit data in real time. The buoys measure things like water temperature, wave movement, and cloudiness, each of which has a hand in high E. coli levels.

## GOAL:

The main goal of this project is to minimize the closure periods due to the time lags between sampling and testing. Using data science techniques, I aim to cut down this time using readings off the buoys and weather stations, and using these readings to presict the bacterial levels of beach water (Beach water quality).

### Data Understanding

The next step is to load the raw datasets. We have the raw .csv files in a 'data' folder in our repository. We load the .csv files using pandas and transform to dataframes for ease of processing. The dataset are all open-sourced and hosted on the City of Chicago Parks and Recreation portal. All observations are readings from weather stations and buoys offshore. There are two datasets I am deploying: the first from the buoys located at the beaches records water quality, or turbidity. (Turbidity is the measure of relative clarity of a liquid. It is an optical characteristic of water and is a measurement of the amount of light that is scattered by material in the water when a light is shined through the water sample. The higher the intensity of scattered light, the higher the turbidity. Material that causes water to be turbid include clay, silt, very tiny inorganic and organic matter, algae, dissolved colored organic compounds, and plankton and other microscopic organisms.)

The second dataset contains observations read of weather sensors located at the beaches. 

The target variable is water quality, indicated as turbidity.

Ocean wave height (technically referred to as amplitude) is a measurement of the vertical distance of the wave from the average height of the wave. The wave axis is the still water level of the ocean, and is usually equated to zero. Wave height above and below the average are measured as positive and negative values, respectively. Could that be the reason we have negative values in the 'Wave Height" and "Wave Period" columns?

However, on closer inspection, notice these two columns have extreme-looking values as their minimum values. These could be attributed to clerical errors, or perhaps filling in for unknown values as it was done several years ago (before the advent of NaNs).

However, it does seem like the above -99999.92 and -100000.0 were arbitrarily entered as values in the above columns for some odd reason.

## Binning months into seasons

For ease of modeling, I am binning the monthly time periods into seasons. The readings are taken hourly mostly during the beach season. The beaches are officially open from the Friday before Memorial Day through Labor Day. 


```python
# Creating monthly variable
turb_df['Month'] = turb_df['Measurement Timestamp'].apply(lambda time: time.month)
```

```python
# extract month from measurement timestamp variable and display number of occurences of each in dataset
turb_df['Measurement Timestamp'].dt.month.value_counts()
```

```python
# display distribution of months
turb_df.Month.value_counts().plot.bar();
```

```python
# Bin timestamp into annual seasons
turb_df['date_offset'] = (turb_df['Measurement Timestamp'].dt.month*100 + 
                          turb_df['Measurement Timestamp'].dt.day - 320)%1300

turb_df['season'] = pd.cut(turb_df['date_offset'], [0, 300, 602, 900, 1300], 
                      labels=['spring', 'summer', 'autumn', 'winter'],include_lowest=True)

# credit: https://stackoverflow.com/a/60285720
```

```python
# display distribution of seasons in dataset
turb_df.season.value_counts().plot.bar();
```

```python
# Display distribution of features
turb_df.hist(figsize=(15,8),bins=10,grid=False);
```

## Binning turbidity level

I am also binning the turbidity readings into turbidity levels, using modified standardized levels of turbidty (Lelykesehatan, Turbidity of Water, WordPress.Com, 2011.)

```python
# Define bins
bins = [0, 5, 35, 50, 2000]
```

```python
# Define the corresponding category for each bin
Levels = ['L', 'M', 'H', 'VH']
```

```python
# create a new column mapping water turbidity to a turbidity level
turb_df['Turbidity_level'] = pd.cut(turb_df['Turbidity'],bins,labels=Levels,ordered=True,include_lowest=True)
```

>>## Second Model: Logistic Regression

For the second logistic regression model, I will be adding in the categorical variables(using sklearn's OneHotEncoder)  and addressing the imbalance in target variable classes with imblearn SMOTE. 

```python
# Create independent variable
drop_list = ['Measurement Timestamp','Turbidity', 'Interval Rain', 'Turbidity_level',
            'date_offset']
X = turb_df.drop(columns=drop_list)

# Create dependent variable
y = turb_df['Turbidity_level']
```

```python
# train-test split of variables
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=42)
```

```python
# Encode categorical variables

object_cols = ['Beach Name','season', 'Station Name']

# Apply one-hot encoder to each column with categorical data
enc_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc_cols_train = pd.DataFrame(enc_encoder.fit_transform(X_train[object_cols]))
enc_cols_test = pd.DataFrame(enc_encoder.transform(X_test[object_cols]))

# One-hot encoding removed index; put it back
enc_cols_train.index = X_train.index
enc_cols_test.index = X_test.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(object_cols, axis=1)
num_X_test = X_test.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
X_train_enc = pd.concat([num_X_train, enc_cols_train], axis=1)
X_test_enc = pd.concat([num_X_test, enc_cols_test], axis=1)
```

### Addressing the imbalance in target variable using SMOTE

```python
# oversample the minority classes of the dependent variable

smote = SMOTE(random_state=2022, sampling_strategy='auto')

X_train_resampled, y_train_resampled = smote.fit_resample(X_train_enc, y_train) 
# Preview synthetic sample class distribution
print('Synthetic sample class distribution: \n')
print(y_train_resampled.value_counts())
```

```python
print(y_train_resampled.shape)
print(X_train_resampled.shape)
```

I will be using a logistic regression instance with a random_state=21 for ease of reproducibility, class_weight = 'balanced' because the target variable class weight is inherently imbalanced, solver='saga' because this solver algorithm handles multiclass problems quite well.

```python
# Initialize logistic regression instance with random_state=21,class_weight='balanced',penalty='L1',
# solver='saga', and max_iter=1000
log = LogisticRegression(random_state=21,class_weight='balanced',max_iter=1000,solver='saga',penalty='l1')
```

```python
# fit instance to training data
log.fit(X_train_resampled,y_train_resampled)
```

## Second Model Evaluation

As stated above, I will be evaluating this model on its accuracy. 

```python
# print accuracy score on training data
print(accuracy_score(y_train_resampled, log.predict(X_train_resampled)))
```

```python
# Print accuracy score on test data
print(accuracy_score(y_test, log.predict(X_test_enc)))
```

This second model has an accuracy less than the first simple logistic regression model. At c.76% accuracy on test data, it does not perform badly.

### Preprocessing Steps

In addition to the processing carried out on the data in preparation for running the second logistic model, this time we are standard scaling the numerical variables. 

```python
# Create independent variable
drop_list2 = ['Measurement Timestamp','Turbidity', 'Interval Rain', 'Turbidity_level',
            'date_offset']
X = turb_df.drop(columns=drop_list2)

# Create dependent variable
y = turb_df['Turbidity_level']
```

```python
# train-test split of variables
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.30,random_state=2022)
```

```python
# OnehotEncode categorical variables
object_cols = ['Beach Name','season', 'Station Name']

# Apply one-hot encoder to each column with categorical data
enc_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc_cols_train2 = pd.DataFrame(enc_encoder.fit_transform(X_train2[object_cols]))
enc_cols_test2 = pd.DataFrame(enc_encoder.transform(X_test2[object_cols]))

# One-hot encoding removed index; put it back
enc_cols_train2.index = X_train2.index
enc_cols_test2.index = X_test2.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train2 = X_train2.drop(object_cols, axis=1)
num_X_test2 = X_test2.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
X_train_enc2 = pd.concat([num_X_train2, enc_cols_train2], axis=1)
X_test_enc2 = pd.concat([num_X_test2, enc_cols_test2], axis=1)
```

```python
X_train_enc2.head()
```

```python
# standard scale numerical features
scaler = StandardScaler()

ss_list = ['Water Temperature','Air Temperature','Wet Bulb Temperature',
           'Humidity','Rain Intensity','Total Rain',
          'Wave Height', 'Wave Period']

scaler.fit(X_train_enc2[ss_list])
```

```python
def scale_values(X, scaler):
    """
    Given a DataFrame and a fitted scaler, use the scaler to scale some of the features
    """
    scaled_array = scaler.transform(X)
    scaled_df = pd.DataFrame(scaled_array, columns=X.columns, index=X.index)
    
    
    return scaled_df

# code credit: Greg Damico
```

```python
scaled_train_df = scale_values(X_train_enc2[ss_list], scaler)
scaled_test_df = scale_values(X_test_enc2[ss_list],scaler)
```

```python
# Concat training dataframe with scaled df
X_train_enc2 = X_train_enc2.drop(ss_list, axis=1)
X_train_enc2 = pd.concat([X_train_enc2, scaled_train_df], axis=1)

# Concat test dataframe with scaled df
X_test_enc2 = X_test_enc2.drop(ss_list, axis=1)
X_test_enc2 = pd.concat([X_test_enc2, scaled_test_df], axis=1)
```

```python
X_train_enc2.head()
```

```python
# oversample the minority classes of the dependent variable

smote = SMOTE(random_state=42, sampling_strategy='minority')

X_train2_resampled, y_train2_resampled = smote.fit_resample(X_train_enc2, y_train2) 
# Preview synthetic sample class distribution
print('Synthetic sample class distribution: \n')
print(y_train_resampled.value_counts())
```

### Build & Evaluate Model

```python
# Instantiate logistic regression instance with the following parameters: penalty='l2', solver='newton-cg',
lr2 = LogisticRegression(penalty='l2', solver='newton-cg', )
```

```python
# Fit the instance to training data
lr2.fit(X_train2_resampled, y_train2_resampled)
```

```python
# Print out accuracy score on training and test data
print(accuracy_score(y_train2_resampled, lr2.predict(X_train2_resampled)))
```

```python
print(accuracy_score(y_test2, lr2.predict(X_test_enc2)))
```

>## 4th Model: Decision Tree Classifier

<br>

On to a fourth and hopefully final model. This time, we are using a Decision Tree Classifier. I will streamline this model using a pipeline. In summary, the numerical variables will be standard scaled, the optimal decision tree parameters will be gotten by using gridsearch, before the usual fitting and transformation. 

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([('ss', StandardScaler()),
                 ('tree', DecisionTreeClassifier(random_state=2022))])
```

```python
grid = [{'tree__max_depth': [None, 2, 6, 10], 
         'tree__min_samples_split': [5, 10]}]
```

```python
gridsearch = GridSearchCV(estimator=pipe, 
                          param_grid=grid, 
                          scoring='accuracy', 
                          cv=7)
```

```python
pipe.fit(X_train2_resampled, y_train2_resampled)
```

```python
pipe.score(X_test_enc2, y_test2)
```

With a Decision Tree Classifier, our model attains a prediction accuracy of c.96% on test data, which is rather impressive.
