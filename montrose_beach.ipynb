```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
```

```python
df = pd.read_csv('~/Downloads/bwq.csv',parse_dates=True,index_col='Measurement Timestamp')
```

```python
fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(15, 8)

sns.relplot(y='Turbidity',x='Water Temperature',data=df, hue='Beach Name');
```

```python
df = df.drop(columns='Transducer Depth').dropna()
```

```python
montrose_df = df[df['Beach Name'] == 'Montrose Beach']
montrose_df
```

```python
montrose_df.plot(x='Turbidity', style='.b', figsize=(15,8));
```

```python
montrose_ts = montrose_df['Turbidity']

montrose_ts
```

```python
montrose_ts.plot(figsize=(12,8))
```

There looks to be an obvious spike in turbidity between months 5 and 9 every year: the height of the Chicago rainy season?

```python
montrose_ts['2014/01':'2014/09'].plot()
```

```python
montrose_ts_diff = montrose_ts.diff(periods=12)
```
